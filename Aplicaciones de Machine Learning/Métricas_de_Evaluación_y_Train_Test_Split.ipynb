# Métricas de Evaluación en Machine Learning

---

## 1. ¿Por qué necesitamos métricas?

Cuando entrenamos un modelo supervisado, siempre nos preguntaremos

**¿Qué tan bien predice nuestro modelo?**

Las métricas de evaluación permiten responder esta pregunta de manera cuantitativa.  
Comparan las **predicciones** ($\hat{y}$) con los **valores reales** ($y$) y nos dan una medida objetiva del rendimiento.

---

## 2. Tipos de métricas según el tipo de problema

| Tipo de problema | Qué se predice | Ejemplo | Tipo de métrica |
|------------------|----------------|----------|-----------------|
| Clasificación | Categorías o etiquetas | Diagnóstico: sano / enfermo | Tasas de acierto o error |
| Regresión | Valores numéricos continuos | Precio de una casa, temperatura | Errores o varianza explicada |

Cada familia de problemas tiene sus propias métricas,  
aunque todas buscan medir **qué tan cerca están las predicciones de la realidad**.

---

## 3. Métricas en Clasificación

En clasificación, el modelo predice **clases** (por ejemplo, 0 o 1, “benigno” o “maligno”).  
Las métricas evalúan cuántas veces el modelo acierta y cómo se comporta con cada clase.

---

### Ejemplo teórico

Un estudio investigó 1250 hombres con sospecha de cáncer de próstata, a los cuales se les realizó una biopsia.  
Los resultados fueron los siguientes:

- 920 hombres tuvieron **biopsia positiva**.  
- 330 hombres tuvieron **biopsia negativa**.  
- Al finalizar el seguimiento se confirmó que de los que dieron positivo, **750 hombres realmente tenían cáncer de próstata**.  
- Además, la prueba tuvo **40 resultados falsos negativos** (personas con cáncer cuya biopsia dio negativa).

Podemos organizar esta información en una matriz de confusión.

---

### Matriz de confusión

|                   | **Predicho: Positivo** | **Predicho: Negativo** |
|-------------------|------------------------|------------------------|
| **Real: Positivo** | Verdadero Positivo (TP) = 750 | Falso Negativo (FN) = 40 |
| **Real: Negativo** | Falso Positivo (FP) = 170 | Verdadero Negativo (TN) = 290 |

---

### a) Accuracy (Exactitud)

El **accuracy** o **exactitud global** mide la proporción de observaciones correctamente clasificadas:

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

En el ejemplo:

$$
\text{Accuracy} = \frac{750 + 290}{750 + 290 + 170 + 40} = \frac{1040}{1250} = 0.832
$$


**Interpretación:** el modelo (o la prueba médica) clasifica correctamente el 83.2% de los casos.



### Matriz de confusión

|                   | **Predicho: Positivo** | **Predicho: Negativo** |
|-------------------|------------------------|------------------------|
| **Real: Positivo** | Verdadero Positivo (TP) = 750 | Falso Negativo (FN) = 40 |
| **Real: Negativo** | Falso Positivo (FP) = 170 | Verdadero Negativo (TN) = 290 |

---

### b) Precisión (Precision)

Mide qué proporción de los casos **predichos como positivos** son realmente positivos:

$$
\text{Precisión} = \frac{TP}{TP + FP}
$$

En el ejemplo:

$$
\text{Precisión} = \frac{750}{750 + 170} = 0.815
$$



### Matriz de confusión

|                   | **Predicho: Positivo** | **Predicho: Negativo** |
|-------------------|------------------------|------------------------|
| **Real: Positivo** | Verdadero Positivo (TP) = 750 | Falso Negativo (FN) = 40 |
| **Real: Negativo** | Falso Positivo (FP) = 170 | Verdadero Negativo (TN) = 290 |

---

### c) Recall (Sensibilidad)

También llamado **Tasa de Verdaderos Positivos (TPR)** o **Sensibilidad**, mide la capacidad del modelo para detectar correctamente los casos positivos reales:

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

En el ejemplo:

$$
\text{Recall} = \frac{750}{750 + 40} = 0.949
$$

### Matriz de confusión

|                   | **Predicho: Positivo** | **Predicho: Negativo** |
|-------------------|------------------------|------------------------|
| **Real: Positivo** | Verdadero Positivo (TP) = 750 | Falso Negativo (FN) = 40 |
| **Real: Negativo** | Falso Positivo (FP) = 170 | Verdadero Negativo (TN) = 290 |

---


### d) Especificidad (Tasa de Verdaderos Negativos)

La **especificidad** mide la proporción de negativos reales correctamente identificados:

$$
\text{Especificidad} = \frac{TN}{TN + FP}
$$

En el ejemplo:

$$
\text{Especificidad} = \frac{290}{290 + 170} = 0.630
$$
### Matriz de confusión

|                   | **Predicho: Positivo** | **Predicho: Negativo** |
|-------------------|------------------------|------------------------|
| **Real: Positivo** | Verdadero Positivo (TP) = 750 | Falso Negativo (FN) = 40 |
| **Real: Negativo** | Falso Positivo (FP) = 170 | Verdadero Negativo (TN) = 290 |

---

### e) F1-score

El **F1-score** combina precisión y recall en una sola métrica armónica:

$$
F1 = 2 \times \frac{\text{Precisión} \times \text{Recall}}{\text{Precisión} + \text{Recall}}
$$

En el ejemplo:

$$
F1 = 2 \times \frac{0.815 \times 0.949}{0.815 + 0.949} = 0.877
$$
### Matriz de confusión

|                   | **Predicho: Positivo** | **Predicho: Negativo** |
|-------------------|------------------------|------------------------|
| **Real: Positivo** | Verdadero Positivo (TP) = 750 | Falso Negativo (FN) = 40 |
| **Real: Negativo** | Falso Positivo (FP) = 170 | Verdadero Negativo (TN) = 290 |

---






Resumen de métricas de clasificación

| Métrica | Fórmula | Qué mide | Ideal |
|----------|----------|-----------|--------|
| Accuracy | $$\frac{TP + TN}{TP + TN + FP + FN}$$ | Exactitud global | 1 |
| Precisión | $$\frac{TP}{TP + FP}$$ | Fiabilidad de positivos | 1 |
| Recall | $$\frac{TP}{TP + FN}$$ | Capacidad de detección | 1 |
| Especificidad | $$\frac{TN}{TN + FP}$$ | Detección correcta de negativos | 1 |
| F1-score | $$2 \times \frac{\text{Precisión} \times \text{Recall}}{\text{Precisión} + \text{Recall}}$$ | Balance P/R | 1 |

#Curva ROC y AUC

En lugar de predecir directamente “positivo” o “negativo”, muchos modelos devuelven **probabilidades**.  
Variando el **umbral de decisión** (por ejemplo, 0.5, 0.6, 0.7...), se obtienen diferentes combinaciones de sensibilidad (TPR) y tasa de falsos positivos (FPR).

- **Curva ROC:** gráfica de TPR (Recall) vs. FPR.
- **AUC (Área Bajo la Curva):** mide el área total bajo la curva ROC.

| AUC | Interpretación |
|-----|----------------|
| 0.5 | Modelo aleatorio |
| 0.7–0.8 | Aceptable |
| 0.8–0.9 | Buena |
| > 0.9 | Excelente |

Cuanto mayor sea el AUC, mejor será el poder de discriminación del modelo.

```python
# ==============================================
# MÉTRICAS BÁSICAS Y MATRIZ DE CONFUSIÓN
# ==============================================

from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
data = load_breast_cancer(as_frame=True)
X = data.frame.drop(columns="target")
y = data.frame["target"]
```

```python
data.target_names
```

```python
X
```

```python
model = LogisticRegression(solver="liblinear")
model.fit(X, y)
```

```python
y_pred = model.predict(X)
y_pred
```

```python
cm = confusion_matrix(y, y_pred)
cm
```

```python
plt.figure(figsize=(12,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Matriz de Confusión")
plt.xlabel("Predicción")
plt.ylabel("Valor Real")
plt.show()
```

```python
print("Reporte de clasificación")
print(classification_report(y, y_pred, target_names=data.target_names))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
# 3. Probabilidades predichas (necesarias para la curva ROC)
y_prob = model.predict_proba(X)[:, 1]   # probabilidad de clase positiva (1 = benigno)

# 4. Calcular puntos de la curva ROC
fpr, tpr, thresholds = roc_curve(y, y_prob)

# 5. Calcular el área bajo la curva (AUC)
auc = roc_auc_score(y, y_prob)

# 6. Graficar la curva ROC
plt.figure(figsize=(5,4))
plt.plot(fpr, tpr, color='blue', label=f'Curva ROC (AUC = {auc:.3f})')
plt.plot([0,1], [0,1], color='red', linestyle='--', label='Azar (AUC = 0.5)')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC - Clasificación binaria (Cáncer de mama)')
plt.legend()
plt.show()

print(f"Área bajo la curva (AUC): {auc:.3f}")
```

## Métricas en Regresión

En regresión, el modelo predice **valores numéricos continuos**.  
Las métricas evalúan cuánto se equivoca el modelo, es decir, la distancia entre el valor real y el predicho.

---

### a) Error Absoluto Medio (MAE)

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

Mide el error promedio en valor absoluto.  
Tiene las mismas unidades que la variable objetivo.  
Ejemplo: si MAE = 5000, el modelo se equivoca en promedio 5000 unidades monetarias.

---

### b) Error Cuadrático Medio (MSE) y Raíz del MSE (RMSE)

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

$$
\text{RMSE} = \sqrt{\text{MSE}}
$$

- Penalizan más los errores grandes (por el cuadrado).  
- RMSE vuelve a las unidades originales, por lo que es más interpretable.

---

### c) Coeficiente de Determinación (R²)

$$
R^2 = 1 - \frac{\sum_i (y_i - \hat{y}_i)^2}{\sum_i (y_i - \bar{y})^2}
$$

| Valor | Interpretación |
|--------|----------------|
| 1 | Predicción perfecta |
| 0 | Igual que predecir el promedio |
| < 0 | Peor que el promedio |

Ejemplo: si $R^2 = 0.84$, el modelo explica el 84% de la variabilidad de los datos.

---

### d) Resumen general (regresión)

| Métrica | Qué mide | Ideal | Comentario |
|----------|-----------|--------|-------------|
| MAE | Error promedio absoluto | 0 | Intuitivo y robusto |
| MSE | Error cuadrático medio | 0 | Penaliza grandes errores |
| RMSE | Raíz del MSE | 0 | Misma unidad que la variable |
| R² | Varianza explicada | 1 | Poder explicativo global |
